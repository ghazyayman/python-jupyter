```python
import math
import random
import matplotlib.pyplot as plt

N = 1000000
a = 0
b = 2
def function(x):
    return x*x

count = 0.0 
in_area = 0.0
count_curv = 0.0

x_up = []
y_up = []

x_down = []
y_down = []

while count < N:
    x_coord = random.uniform(0,2)
    y_coord = random.uniform(0,4)
    
    if y_coord < function(x_coord):
        in_area += 1
        x_up.append(x_coord)
        y_up.append(y_coord)
      
    else:
        x_down.append(x_coord)
        y_down.append(y_coord)

    count += 1
    
area_box = 2 * 4 
print((in_area/count) * area_box)

plt.plot(x_down, y_down, 'o', c = 'maroon', ms=0.1, alpha=0.5)
plt.plot(x_up, y_up, 'o', c = 'teal', ms=0.1, alpha=0.5)
plt.show()
```

Monte Carlo

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def dice_roll():
    return random.randint(1,6) 

list1 = []

def monte_carlo(n):
    results = 0
    for i in range(n):
        flip_result = dice_roll()
        if flip_result == 2:
          results += 1
          prob_value = results/(i+1)
          list1.append(prob_value)

        # #Plot the results :
        plt.axhline(y=0.167, color='r', linestyle='-')
        plt.xlabel("Iterations")
        plt.ylabel("Probability")
        plt.plot(list1)
       
    return results/n

answer = monte_carlo(1000)
print("Final value :",answer)
```
